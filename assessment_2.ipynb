```python
import random
from itertools import combinations
```

```python
traffic = [random.randint(50, 400) for i in range(20)]
```

```python
traffic
```

```python
link = [(random.randint(50, 400), random.randint(50, 400)) for _ in range(10)]
```

```python
link
```

```python
for idx, jitter_limit in enumerate(traffic):
    max_bandwidth = 0
    best_combo = []

    for r in range(2, 11):
        for combo in combinations(link, r):
            # print(combo)
            latencies = [lat for lat, bw in combo]
            min_lat = min(latencies)
            max_lat = max(latencies)
            jitter = max_lat - min_lat

            if jitter <= jitter_limit:
                bandwidth_sum = sum(bw for lat, bw in combo)
                if bandwidth_sum > max_bandwidth:
                    max_bandwidth = bandwidth_sum
                    best_combo = combo
    


    print(f"Traffic Type {idx+1}:")
    print(f"  - Jitter Limit: {jitter_limit}")
    print(f"  - Bandwidth Allocated: {max_bandwidth}")
    print(f"  - Links Used: {best_combo}\n")
```

```python
link.sort(key=lambda x: x[0])

for t_idx, jitter_limit in enumerate(traffic):
    max_bandwidth = 0
    best_window = []

    start = 0
    bandwidth_sum = 0
    window_links = []

    for end in range(len(link)):
        window_links.append(link[end])
        bandwidth_sum += link[end][1]

        while link[end][0] - link[start][0] > jitter_limit:
            bandwidth_sum -= link[start][1]
            window_links.pop(0)
            start += 1

        if bandwidth_sum > max_bandwidth:
            max_bandwidth = bandwidth_sum
            best_window = window_links.copy()

    print(f"Traffic Type {t_idx+1}:")
    print(f"  - Jitter Limit: {jitter_limit}")
    print(f"  - Bandwidth Allocated: {max_bandwidth}")
    print(f"  - Links Used: {best_window}\n")
```
